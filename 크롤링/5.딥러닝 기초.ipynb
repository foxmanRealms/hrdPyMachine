```python
import pandas as pd
import numpy as np
```

```python
X=[[0,0],
  [0,1],
  [1,0],
  [1,1],]

y=[0,0,0,1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation       ##### 액티베이션   활성화함수
```

```python
# 신경망 생성
model= Sequential()  # 도화지       ##### 도화지를 깐다 그런 느낌   위에서부터 순서대로 들어가게 된다
model.add( Dense(units=1, input_dim=2 )    ) # Dense   : 층을 추가 , input_dim=입력갯수(특성수)   ##### 특성숫자 ~- 컬럼숫자 
model.add(Activation('sigmoid') )  #units : 출력갯수
```

```python
model.summary()  # 신경망 구조 출력
```

```python
# 컴파일 : 설계한 신경망을 활용할 수 있도록 모델로 만들어줌

#손실함수 : loss
#2진회귀 : mse,mean squared error
#2진분류 : binary_crossentropy
#다중분류 : categorial_crossentropy

#경사하강법 종류 : optimizer
#rmsprop
#adam
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=1000) 
```

```python
# 신경망 생성
model= Sequential()  
model.add( Dense(4,input_dim=2 ,activation='relu')    ) # 입력층

model.add(Dense(8 ,activation='relu'))  # 은닉층   ##### 입력층 출력충 사이에 숨어있는 층이고, 그 수가 여러개이어도된다
model.add(Dense(1 ,activation='sigmoid'))  # 출력층
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=5000) 
```

```python
model.predict( [[0,1]])
```
